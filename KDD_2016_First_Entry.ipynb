```python
import pandas as pd
import numpy as np
```

```python
affiliations = pd.read_table("data/2016KDDCupSelectedAffiliations/2016KDDCupSelectedAffiliations.txt",names = ["id","university"])
affiliations.head()
```

```python
selectedPapers = pd.read_table("data/2016KDDCupSelectedPapers/2016KDDCupSelectedPapers.txt",names = ["id","title","year","conferenceId","conferenceShortName"])
selectedPapers.head()
```

```python
pd.unique(selectedPapers.conferenceShortName.ravel())
```

```python
sigirContestPapers = selectedPapers.loc[selectedPapers['conferenceShortName'].isin(["SIGIR"])]
sigirContestPapers.head()
```

```python
sigirPaperIds = pd.unique(sigirContestPapers.id.ravel())
sigirPaperIds[:20]
```

```python
contestPapersSigmod = selectedPapers.loc[selectedPapers['conferenceShortName'].isin(["SIGMOD"])]
contestPapersSigmod.head()
```

```python
sigmodPaperIds = pd.unique(contestPapersSigmod.id.ravel())
sigmodPaperIds[:20]
```

```python
contestPapersSigcom = selectedPapers.loc[selectedPapers['conferenceShortName'].isin(["SIGCOMM"])]
contestPapersSigcom.head()
```

```python
sigcomPaperIds = pd.unique(contestPapersSigcom.id.ravel())
sigcomPaperIds[:20]
```

```python
pafReader = pd.read_table("data/PaperAuthorAffiliations/PaperAuthorAffiliations.txt",chunksize=1000000,names = ["paperId","authorId","affiliationId","origAffilName","normAffilName","authorSeqNum"],dtype={"paperId":str,"authorId":str,"affiliationId":str,"origAffilName":str,"normAffilName":str,"authorSeqNum":np.int32})
```

```python
chunk = pafReader.get_chunk()
sigirPaperAuths = chunk[chunk.paperId.isin(sigirPaperIds)]
sigmodPaperAuths = chunk[chunk.paperId.isin(sigmodPaperIds)]
sigcomPaperAuths = chunk[chunk.paperId.isin(sigcomPaperIds)]
for chunk in pafReader:
    sgrpa = chunk[chunk.paperId.isin(sigirPaperIds)]
    smdpa = chunk[chunk.paperId.isin(sigmodPaperIds)]
    scmpa = chunk[chunk.paperId.isin(sigcomPaperIds)]
    
    sigirPaperAuths = pd.concat([sigirPaperAuths,sgrpa])
    sigmodpaperAuths = pd.concat([sigmodPaperAuths,smdpa])
    sigcomPaperAuths = pd.concat([sigcomPaperAuths,scmpa])
    print sigirPaperAuths.shape , sigmodpaperAuths.shape , sigcomPaperAuths.shape
```

```python
sigirPaperAuths = pd.read_csv("data/sigirPaperAuthorAffilitaions")
```

```python
spaTop = sigirPaperAuths.groupby('affiliationId').count().sort(["paperId"],ascending=False)
```

```python
sigirTopAffil = spaTop[spaTop["paperId"] > 5]
```

```python
sigirTopAffil.head()
```

```python
sigmodpa = pd.read_csv("data/sigmodPaperAuthorAffiliations")
```

```python
sigmodTop = sigmodpa.groupby('affiliationId').count().sort(["paperId"],ascending=False)
```

```python
sigmodTopAffil = sigmodTop[sigmodTop["paperId"] > 5]
```

```python
sigmodTopAffil.head()
```

```python
sigcompa = pd.read_csv("data/sigcomPaperAuthorAffiliations")
```

```python
sigcomTop = sigcompa.groupby('affiliationId').count().sort(["paperId"],ascending=False)
```

```python
sigcomTopAffil = sigcomTop[sigcomTop["paperId"]>5]
```

```python
sigcomTopAffil.head()
```

```python
sigirValues = [[]]
for i in sigirTopAffil.index:
    sigirValues.append([i,sigirTopAffil.loc[i].paperId])
sigirValues.remove([])
sigirValues = np.array(sigirValues)
sum = sigirValues[:,1].astype(int).sum()
sigirWeight = [[]]
for i in range(0,sigirValues.size/2):
    sigirWeight.append(["43FD776C",sigirValues[i,0],sigirValues[i,1].astype(float)/sum])
sigirWeight.remove([])
```

```python
sigirWeight[:10]
```

```python
sigmodValues = [[]]
for i in sigmodTopAffil.index:
    sigmodValues.append([i,sigmodTopAffil.loc[i].paperId])
sigmodValues.remove([])
sigmodValues = np.array(sigmodValues)
sigmodsum = sigmodValues[:,1].astype(int).sum()
sigmodWeight = [[]]
for i in range(0,sigmodValues.size/2):
    sigmodWeight.append(["460A7036",sigmodValues[i,0],sigmodValues[i,1].astype(float)/sum])
sigmodWeight.remove([])
```

```python
sigcomValues = [[]]
for i in sigcomTopAffil.index:
    sigcomValues.append([i,sigcomTopAffil.loc[i].paperId])
sigcomValues.remove([])
sigcomValues = np.array(sigcomValues)
sigcomsum = sigmodValues[:,1].astype(int).sum()
sigcomWeight = [[]]
for i in range(0,sigcomValues.size/2):
    sigcomWeight.append(["44B13001",sigcomValues[i,0],sigcomValues[i,1].astype(float)/sum])
sigcomWeight.remove([])
```

```python
submission1 = np.concatenate((sigirWeight,sigcomWeight,sigmodWeight),axis=0)
```

```python
subdf = pd.DataFrame(submission1)
```

```python
subdf.columns = ["confId","affilId","prob"]
mask = subdf["affilId"].isin(affiliations["id"])
result = subdf[mask]
result
```

```python
result.to_csv("submissions/results.tsv",sep = "\t",index = False,header = False)
```
