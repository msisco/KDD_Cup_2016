Webcrawl of Urls of sigir website

```python
sigir2016webUrls = set([])
```

```python
import re, urllib

textfile = file('sigir2016.txt','wt')
for i in re.findall('''href=["'](http:\/\/sigir\.org\/sigir2016\/[^wp|x].[^"']+)["']''', urllib.urlopen("http://sigir.org/sigir2016/").read(), re.I):
        sigir2016webUrls.add(i) 
        for ee in re.findall('''href=["'](http:\/\/sigir\.org\/sigir2016\/[^wp|x].[^"']+)["']''', urllib.urlopen(i).read(), re.I):
            textfile.write(ee+'\n')
textfile.close()
```
